```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline    
  #grafikleri daha iyi görmemizi
```

```python
kiloListesi = np.random.randint(30,70,10) #30-70 arasında rastgele 10 sayı yazıyoruz
```

```python
kiloListesi
```

```python
yasListesi = np.arange(10,110,10) # 10-110 arasını 10 atlayarak yazar
```

```python
yasListesi
```

```python
numpykiloListesi = np.array(kiloListesi)
numpyyasListesi = np.array(yasListesi)
```

```python
numpykiloListesi
```

```python
numpyyasListesi
```

```python
plt.plot(numpyyasListesi,numpykiloListesi,"b")  #"b"= blue


plt.xlabel("Yas Listesi")
plt.ylabel("Kilo Listesi")
plt.title("Kilonun Yasa Oranı")

#plt.show()
```

```python
yeniDizi = np.arange(10,80,10)
```

```python
yeniDizi
```

```python
number = np.linspace(10,30,5)
```

```python
number
```

```python
numpyDizisi = yeniDizi ** 2
```

```python
numpyDizisi
```

```python
plt.plot(yeniDizi,numpyDizisi,"g--")
```

```python
plt.plot(yeniDizi,numpyDizisi,"y*-")
```

```python
plt.subplot(1,2,1) 
plt.plot(yeniDizi,numpyDizisi,"r*-")

plt.subplot(1,2,2) # 1 sıra olacak,2 kolonolacak, şu an 2.grafik çiziliyor.
plt.plot(numpyDizisi,yeniDizi,"b--")
```

```python
benimFigure = plt.figure()

figureAxes = benimFigure.add_axes([0.1,0.1,0.9,0.9]) #width,high=genişlik-yükseklik
figureAxes.plot(yeniDizi,numpyDizisi,"y")

figureAxes.set_xlabel("X ekseni İsmi")
figureAxes.set_ylabel("Y ekseni ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3]) #grafiğin yeri için

eksen1.plot(yeniDizi,numpyDizisi,"b")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Ana Grafik başlık")

eksen2.plot(numpyDizisi,yeniDizi,"r")
eksen2.set_xlabel("x ekseni küçük")
eksen2.set_ylabel("y ekseni küçük")
eksen2.set_title("Küçük Grafik başlık")

```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1,ncols=2)  #nrows,ncols:n satırlar,n sütun
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpykiloListesi,numpyyasListesi,"y")
    eksen.set_xlabel("X Ekseni")

plt.tight_layout()  #iki grafik arasındaki mesafeyi açar
benimFigure
```

```python
type(benimFigure)
```

# Görselleri Geliştirme

```python
yeniFigur = plt.figure(dpi = 100)  #dpi: çözünürlük için
 
#plt.figure(figsize = (5,5))=boyutlandırma için kullanılabilir

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyyasListesi, numpyyasListesi**2, label="numpyDizisi**2")
yeniEksen.plot(numpyyasListesi, numpyyasListesi**3, label="numpylist **3")

#label="numpyDizisi**2 :veri açıklaması için

yeniEksen.legend(loc=2)
```

```python
yeniFigur.savefig("benimFigur.png",dpi=100)  #grafiği kaydetme
```
